```python
def gd_points(grad_f, x0, nabla, n_steps):
    points = [x0]
    x = x0
    for _ in range(n_steps):
        grad = grad_f(x)
        x = x - nabla * grad
        points.append(x)
    return points
```

```python
import matplotlib.pyplot as plt
import numpy as np
from gofi.plot.colors import * 

def f(x):
    return np.sin(x)*np.sqrt(x+2)

def grad_f(x):
    return np.cos(x)*np.sqrt(x+2) + np.sin(x)/(2*np.sqrt(x+2))

x = np.linspace(0, 12, 120)
y = f(x)

n_steps= 20
points1 = gd_points(grad_f, x0=2, nabla=0.1, n_steps=n_steps)
points2 = gd_points(grad_f, x0=6, nabla=0.1, n_steps=n_steps)
points3 = gd_points(grad_f, x0=9, nabla=0.1, n_steps=n_steps)


plt.axis('off')
plt.plot(x, y, color=blue, linewidth=2.15)
plt.plot(points1[0], f(np.array(points1[0])),  color=orange, marker='x', alpha=1, linestyle='None', markersize=7)
plt.plot(points1[1:], f(np.array(points1[1:])),  color=orange, marker='.', alpha=1, linestyle='None', markersize=7)
#plt.plot(points2, f(np.array(points2)),  color=yellow, marker='.', alpha=0.7, linestyle='None', markersize=7)

plt.plot(points3[0], f(np.array(points3[0])),  color=pink, marker='x', alpha=1, linestyle='None', markersize=7)
plt.plot(points3[1:], f(np.array(points3[1:])),  color=pink, marker='.', alpha=1, linestyle='None', markersize=7)
plt.savefig('gradientni_spust_primer.pdf', bbox_inches='tight', pad_inches=0, transparent=True)
#plt.axes
```
