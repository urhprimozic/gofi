```python
from sympy.combinatorics.fp_groups import FpGroup
from sympy.combinatorics.free_groups import free_group


# dihedral group:
def D2n(n: int):
    F, s, r = free_group("s, r")
    relators = [s**2, r**n, r * s * r * s]
    return FpGroup(F, relators)
```
